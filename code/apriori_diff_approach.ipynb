```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv('../archive/ratings_small.csv')
ratings.head()
```

```python
movies = pd.read_csv('../archive/movies_metadata.csv')
nan = movies['title'].isna()
movies = movies.loc[nan == False]
```

```python
movies = movies.astype({'id': 'int64'})
```

```python
final = pd.merge(ratings, movies[['id', 'title']], left_on='movieId', right_on='id')

final.head()
```

```python
final.drop(['timestamp', 'id'], axis = 1, inplace=True)
```

```python
final.head()
```

```python
print(final)
```

```python
final = final.drop_duplicates(['userId', 'title'])
```

```python
final = final.pivot(index='userId', columns='title', values = 'rating').fillna(0)
final.head()
```

```python
final = final.astype('int64')
def encode(x):
    if x<=0:
        return 0
    if x>=1:
        return 1

final = final.applymap(encode)

final.head()
```

```python
from mlxtend.frequent_patterns import apriori
freq = apriori(final, min_support = 0.07, use_colnames = True)
freq.head()
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq, metric='lift', min_threshold=1)
rules.head()
```

```python
res =  rules.sort_values(by=['lift'], ascending=False)
res.head()
```

```python
check = res[res['antecedents'].apply(lambda x: len(x)==1 and next(iter(x))=='Back to the Future Part II')]
check = check[check['lift']>2]
check.head()
```

```python
mv = check['consequents'].values
movie_list = []
for movie in mv:
    for title in movie:
        if title not in movie_list:
            movie_list.append(title)

print(movie_list)
```
